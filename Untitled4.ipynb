```python
sede = input("Você está com sede? (sim/nao): ")

if sede == "sim":
    ir = input("Quer ir até a geladeira? (sim/nao): ")

    if ir == "sim":
        abrir = input("Quer abrir a geladeira? (sim/nao): ")

        if abrir == "sim":
            agua = input("Tem água na geladeira? (sim/nao): ")

            if agua == "sim":
                beber = input("Quer beber água? (sim/nao): ")

                if beber == "sim":
                    print("Você bebeu água")
                else:
                    print("Então continua com sede")

            else:
                print("Não tem água")

        else:
            print("Então não abriu a geladeira")

    else:
        print("Então continua com sede parado aí")

else:
    print("Então tá tranquilo")
```
